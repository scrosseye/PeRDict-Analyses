```python
import pandas as pd

import spacy
proc = spacy.load("en_core_web_sm")
```

```python
#import texts
df = pd.read_csv('clear_corpus_final.csv') #final texts
df
```

```python
#call in dataframe and convert to dictionary

rhymes = pd.read_csv ('full_data_elp_perfect_rhymes.csv')

print(rhymes)

print(rhymes.size)
print(rhymes.shape)
```

```python
#reads in rhymes, sets key to word, transforms data, changes to dictionary
rhymes_dic2 = rhymes.set_index('Word').T.to_dict('list') #make dictionary key and values (as a list). This makes it easier to call values in slices

#df.set_index('ID').T.to_dict('list')
```

```python
#what does the dictionary look like?

#for key, val in rhymes_dic2.items():
#  if key == "fast":
#    print(key)
#    print(val)

#for key, val in rhymes_dic2.items():
#  if key == "fast":
#    print(key)
#    print(val[1])



c = 0

for key, val in rhymes_dic2.items():
  print(key)
  print(val)
  c += 1
  if c > 15:
    break
```

THIS IS THE MAIN CODE FOR UNLEMMATIZED DATA

```python

cw_length = []
num_rhymes_full_elp_corp = []
num_rhymes_1000_coca_corp = []
num_rhymes_2500_coca_corp = []
num_rhymes_5000_coca_corp = []
num_rhymes_10000_coca_corp = []



for index, row in df.iterrows():
    text = str(row["Text"])  # Convert text to string for spacy processing
    text = text.lower() #lowercase the stuff
    cw_token = []
    num_rhymes_full_elp = []
    num_rhymes_1000_coca = []
    num_rhymes_2500_coca = []
    num_rhymes_5000_coca = []
    num_rhymes_10000_coca = []
#    print(text)
    tokenized_doc = proc(text)
    for token in tokenized_doc:
        #print(token.text)
        if not token.is_stop: #remove stop words
            if token.is_alpha: #remove all non-alpha crap
                #print(token.text)
                cw_token.append(str(token))
                for key, value in rhymes_dic2.items():
                    if key == token.text:
                        #print(key)
                        #print(value[0])
                        num_rhymes_full_elp.append(value[0])
                        num_rhymes_1000_coca.append(value[1])
                        num_rhymes_2500_coca.append(value[2])
                        num_rhymes_5000_coca.append(value[3])
                        num_rhymes_10000_coca.append(value[4])
    #print(num_rhymes_full_elp)
    
    cw_length.append(len(cw_token))
    num_rhymes_full_elp_corp.append(num_rhymes_full_elp)
    num_rhymes_1000_coca_corp.append(num_rhymes_1000_coca)
    num_rhymes_2500_coca_corp.append(num_rhymes_2500_coca)
    num_rhymes_5000_coca_corp.append(num_rhymes_5000_coca)
    num_rhymes_10000_coca_corp.append(num_rhymes_10000_coca)

#print(f'this is the length of the texts by content words {cw_length}') 
#print(f'this is a list of full elp rhymes by text {num_rhymes_full_elp_corp}')
#print(f'this is a list of elp rhymes 1000 COCA by text {num_rhymes_1000_coca_corp}')


#get lists that are sums of sublists
rhymes_full_elp = [sum(sub_list) for sub_list in num_rhymes_full_elp_corp]
#print(rhymes_full_elp)
rhymes_1000_coca = [sum(sub_list) for sub_list in num_rhymes_1000_coca_corp]
rhymes_2500_coca = [sum(sub_list) for sub_list in num_rhymes_2500_coca_corp]
rhymes_5000_coca = [sum(sub_list) for sub_list in num_rhymes_5000_coca_corp]
rhymes_10000_coca = [sum(sub_list) for sub_list in num_rhymes_10000_coca_corp]




num_rhymes_full_elp_final = [(a+.00000000000000000000000001)/(b) for a, b in zip(rhymes_full_elp, cw_length)] 
#print(num_rhymes_full_elp_final)
rhymes_1000_coca_final = [(a +.00000000000000000000000001)/(b) for a, b in zip(rhymes_1000_coca, cw_length)] 
#print(rhymes_1000_coca_final)
rhymes_2500_coca_final = [(a +.00000000000000000000000001)/(b) for a, b in zip(rhymes_2500_coca, cw_length)] 
rhymes_5000_coca_final = [(a +.00000000000000000000000001)/(b) for a, b in zip(rhymes_5000_coca, cw_length)] 
rhymes_10000_coca_final = [(a +.00000000000000000000000001)/(b) for a, b in zip(rhymes_10000_coca, cw_length)] 



```

```python
#add to database

df2 = df.assign(num_rhymes_full_elp_final_tokens = num_rhymes_full_elp_final, rhymes_1000_coca_final_tokens = rhymes_1000_coca_final,
               rhymes_2500_coca_final_tokens = rhymes_2500_coca_final, rhymes_5000_coca_final_tokens = rhymes_5000_coca_final,
               rhymes_10000_coca_final_tokens = rhymes_10000_coca_final)

df2

#df2.to_csv("num_rhymes_no_lemmas.csv")
```

THIS IS FOR LEMMATIZED DATA

```python
cw_length_lemma = []
num_rhymes_full_elp_corp_lemma = []
num_rhymes_1000_coca_corp_lemma = []
num_rhymes_2500_coca_corp_lemma = []
num_rhymes_5000_coca_corp_lemma = []
num_rhymes_10000_coca_corp_lemma = []



for index, row in df.iterrows():
    text = str(row["Excerpt"])  # Convert text to string for spacy processing
    text = text.lower() #lowercase the stuff
    cw_token_lemma = []
    num_rhymes_full_elp_lemma = []
    num_rhymes_1000_coca_lemma = []
    num_rhymes_2500_coca_lemma = []
    num_rhymes_5000_coca_lemma = []
    num_rhymes_10000_coca_lemma = []
    #print(text)
    tokenized_doc = proc(text)
    for token in tokenized_doc:
        #print(token.lemma_)
        if not token.is_stop: #remove stop words
            if token.is_alpha: #remove all non-alpha crap
                #print(token.text)
                cw_token_lemma.append(str(token))
                for key, value in rhymes_dic2.items():
                    if key == token.lemma_:
                        #print(key)
                        #print(value[0])
                        #print(value[1])
                        #print(value[2])

                        num_rhymes_full_elp_lemma.append(value[0])
                        num_rhymes_1000_coca_lemma.append(value[1])
                        num_rhymes_2500_coca_lemma.append(value[2])
                        num_rhymes_5000_coca_lemma.append(value[3])
                        num_rhymes_10000_coca_lemma.append(value[4])
    #print(num_rhymes_full_elp)
    
    cw_length_lemma.append(len(cw_token_lemma))
    num_rhymes_full_elp_corp_lemma.append(num_rhymes_full_elp_lemma)
    num_rhymes_1000_coca_corp_lemma.append(num_rhymes_1000_coca_lemma)
    num_rhymes_2500_coca_corp_lemma.append(num_rhymes_2500_coca_lemma)
    num_rhymes_5000_coca_corp_lemma.append(num_rhymes_5000_coca_lemma)
    num_rhymes_10000_coca_corp_lemma.append(num_rhymes_10000_coca_lemma)

#print(f'this is the length of the texts by content words {cw_length_lemma}') 
#print(f'this is a list of full elp rhymes by text {num_rhymes_full_elp_corp_lemma}')
#print(f'this is a list of elp rhymes 1000 COCA by text {num_rhymes_1000_coca_corp_lemma}')
#print(f'this is a list of elp rhymes 2500 COCA by text {num_rhymes_2500_coca_corp_lemma}')


#get lists that are sums of sublists
rhymes_full_elp_lemma = [sum(sub_list) for sub_list in num_rhymes_full_elp_corp_lemma]
#print(rhymes_full_elp_lemma)
rhymes_1000_coca_lemma = [sum(sub_list) for sub_list in num_rhymes_1000_coca_corp_lemma]
rhymes_2500_coca_lemma = [sum(sub_list) for sub_list in num_rhymes_2500_coca_corp_lemma]
rhymes_5000_coca_lemma = [sum(sub_list) for sub_list in num_rhymes_5000_coca_corp_lemma]
rhymes_10000_coca_lemma = [sum(sub_list) for sub_list in num_rhymes_10000_coca_corp_lemma]




num_rhymes_full_elp_final_lemma = [(a+.00000000000000000000000001)/(b) for a, b in zip(rhymes_full_elp_lemma, cw_length_lemma)] 
#print(f'this is the number of rhymes for the entire dataset {num_rhymes_full_elp_final_lemma}')
rhymes_1000_coca_final_lemma = [(a +.00000000000000000000000001)/(b) for a, b in zip(rhymes_1000_coca_lemma, cw_length_lemma)] 
#print(f'this is final rhymes 1000 {rhymes_1000_coca_final_lemma}')
rhymes_2500_coca_final_lemma = [(a +.00000000000000000000000001)/(b) for a, b in zip(rhymes_2500_coca_lemma, cw_length_lemma)] 
#print(f'this is final rhymes 2500 {rhymes_2500_coca_final_lemma}')
rhymes_5000_coca_final_lemma = [(a +.00000000000000000000000001)/(b) for a, b in zip(rhymes_5000_coca_lemma, cw_length_lemma)] 
rhymes_10000_coca_final_lemma = [(a +.00000000000000000000000001)/(b) for a, b in zip(rhymes_10000_coca_lemma, cw_length_lemma)] 

```

```python
#add to database

df3 = df.assign(num_rhymes_full_elp_final_lemma = num_rhymes_full_elp_final_lemma, rhymes_1000_coca_final_lemma = rhymes_1000_coca_final_lemma,
               rhymes_2500_coca_final_lemma = rhymes_2500_coca_final_lemma, rhymes_5000_coca_final_lemma = rhymes_5000_coca_final_lemma,
               rhymes_10000_coca_final_lemma = rhymes_10000_coca_final_lemma)
df3
```
